```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Load the covariance matrix
covariance_matrix = torch.load("/home/aac/ssrivas9/prismatic-vlms/text_covariance_186K.pt")

# Convert to numpy array for visualization
covariance_np = covariance_matrix.cpu().numpy()

# Create a figure with larger size
plt.figure(figsize=(12, 10))

# Create heatmap using matplotlib
im = plt.imshow(covariance_np, aspect='equal')
plt.title('Text Embedding Covariance Matrix')
plt.xlabel('Dimension')
plt.ylabel('Dimension')

# Add colorbar
plt.colorbar(im, label='Covariance')

plt.tight_layout()
plt.show()

# Print some statistics
print(f"Matrix shape: {covariance_np.shape}")
print(f"Min value: {covariance_np.min():.4f}")
print(f"Max value: {covariance_np.max():.4f}")
print(f"Mean value: {covariance_np.mean():.4f}")
print(f"Standard deviation: {covariance_np.std():.4f}")

# Calculate and display eigenvalue statistics
eigenvalues = np.linalg.eigvals(covariance_np)
print("\nEigenvalue statistics:")
print(f"Min eigenvalue: {eigenvalues.min():.4f}")
print(f"Max eigenvalue: {eigenvalues.max():.4f}")
print(f"Condition number: {np.abs(eigenvalues.max() / eigenvalues.min()):.4f}")

```
